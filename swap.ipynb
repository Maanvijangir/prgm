# **Swap with temp variable...**


```python
%%writefile swap.c
#include<stdio.h>
void main()
{
  int a=2,b=3,temp;
  temp=a;
  a=b;
  b=temp;
  printf("a=%d  b=%d",a,b);
}
```

# **Fibonacci Recursion**

```python
%%writefile add.c
#include <stdio.h>

int fib(int n) {
    if (n <= 1) {
        return n;
    } else {
        return fib(n - 1) + fib(n - 2);
    }
}

int main() {
    int n;
    printf("Enter terms: ");
    scanf("%d", &n);

    for (int i = 0; i < n; i++) {
        printf("%d ", fib(i));
    }
    printf("\n");
    return 0;
}
```

```python
!gcc add.c -lm
```

```python
!./a.out
```

# **Transpose of a Matrix**

```python
%%writefile tran.c
#include <stdio.h>

int main() {

    int a[10][10];
    int n, m;


    printf("Enter rows and columns: ");

    scanf("%d%d", &n, &m);


    printf("Enter elements of the matrix:\n");

    for (int i = 0; i < n; i++) {

        for (int j = 0; j < m; j++) {
            scanf("%d", &a[i][j]);
        }
    }


    printf("Transpose of the matrix:\n");

    for (int i = 0; i < m; i++) {

        for (int j = 0; j < n; j++) {
            printf("%d ", a[j][i]);
        }
        printf("\n");
    }
    return 0;
}
```

```python
!gcc tran.c -lm
```

```python
!./a.out
```

# **Operations of Matrix**

```python
%%writefile matrix.c
#include <stdio.h>

int main() {

    int a[10][10], b[10][10], c[10][10];
    int r, c1;
    int ch;


    printf("Enter rows and columns for matrices: ");

    scanf("%d%d", &r, &c1);


    printf("Enter elements of the first matrix:\n");

    for (int i = 0; i < r; i++) {

        for (int j = 0; j < c1; j++) {
            scanf("%d", &a[i][j]); // Read element for matrix 'a'
        }
    }


    printf("Enter elements of the second matrix:\n");

    for (int i = 0; i < r; i++) {

        for (int j = 0; j < c1; j++) {
            scanf("%d", &b[i][j]); // Read element for matrix 'b'
        }
    }


    printf("Choose an operation:\n");
    printf("1. Addition\n");
    printf("2. Subtraction\n");
    printf("3. Multiplication\n");
    printf("Enter your choice (1, 2, or 3): ");
    scanf("%d", &ch);


    if (ch == 1) { // Matrix Addition
        printf("Result of Addition:\n");
        for (int i = 0; i < r; i++) {
            for (int j = 0; j < c1; j++) {
                c[i][j] = a[i][j] + b[i][j];
                printf("%d ", c[i][j]);
            }
            printf("\n");
        }
    } else if (ch == 2) { // Matrix Subtraction
        printf("Result of Subtraction:\n");
        for (int i = 0; i < r; i++) {
            for (int j = 0; j < c1; j++) {
                c[i][j] = a[i][j] - b[i][j];
                printf("%d ", c[i][j]);
            }
            printf("\n");
        }
    } else if (ch == 3) { // Matrix Multiplication
        printf("Result of Multiplication:\n");
        for (int i = 0; i < r; i++) {
            for (int j = 0; j < c1; j++) {
                c[i][j] = 0;
                for (int k = 0; k < c1; k++) {
                    c[i][j] += a[i][k] * b[k][j];
                }
                printf("%d ", c[i][j]);
            }
            printf("\n");
        }
    } else {
        printf("Invalid choice. Please enter 1, 2, or 3.\n"); // Error for invalid choice
    }

    return 0; // Indicate successful program execution
}
```

```python
!gcc matrix.c -lm
```
