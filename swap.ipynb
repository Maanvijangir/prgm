# **Swap with temp variable...**


```python
%%writefile swap.c
#include<stdio.h>
void main()
{
  int a=2,b=3,temp;
  temp=a;
  a=b;
  b=temp;
  printf("a=%d  b=%d",a,b);
}
```

```python
!gcc swap.c -lm
```

```python
!./a.out
```

# **One line swap**

```python
%%writefile swap.c
#include<stdio.h>
void main()
{
  int a=3, b=6;
  a=(a*b)/(b=a);
  printf("a=%d  b=%d",a,b);
}
```

```python
!gcc swap.c -lm
```

```python
!./a.out
```

# **Fibonacci Recursion**

```python
%%writefile add.c
#include <stdio.h>

int fib(int n) {
    if (n <= 1) {
        return n;
    } else {
        return fib(n - 1) + fib(n - 2);
    }
}

int main() {
    int n;
    printf("Enter terms: ");
    scanf("%d", &n);

    for (int i = 0; i < n; i++) {
        printf("%d ", fib(i));
    }
    printf("\n");
    return 0;
}
```

```python
!gcc add.c -lm
```
