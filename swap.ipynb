# **Swap with temp variable...**


```python
%%writefile swap.c
#include<stdio.h>
void main()
{
  int a=2,b=3,temp;
  temp=a;
  a=b;
  b=temp;
  printf("a=%d  b=%d",a,b);
}
```

```python
!gcc swap.c -lm
```

```python
!./a.out
```

# **Swap without temp var...**

```python
%%writefile swap.c
#include<stdio.h>
void main()
{
  int a=1,b=2;
  a=a+b;
  b=a-b;
  a=a-b;
  printf("a=%d  b=%d",a,b);
}
```

```python
!gcc swap.c -lm
```

```python
!./a.out
```

# **One line swap**

```python
%%writefile swap.c
#include<stdio.h>
void main()
{
  int a=3, b=6;
  a=(a*b)/(b=a);
  printf("a=%d  b=%d",a,b);
}
```

```python
!gcc swap.c -lm
```

```python
!./a.out
```
