```python
%%writefile fact.c
#include<stdio.h>
#include<time.h>

int fact_recursive(int n){
  if(n==0  || n==1)
  return 1;
  else
  return n*fact_recursive(n-1);
}

int fact_iterative(int n){
  int fact=1;
  for(int i=1;i<=n;i++){
    fact=fact*i;
  }
  return fact;
}

int main()
{
  int n=6;

  clock_t start_t, end_t;
  double cpu_time_used;

  start_t=clock();
  int rec_result=fact_recursive(n);
  end_t=clock();
  cpu_time_used=((double)(end_t-start_t))/CLOCKS_PER_SEC;
  printf("factorial of %d is %d using recursive method & it take %f seconds..",n,rec_result,cpu_time_used);

  start_t=clock();
  int ite_result=fact_iterative(n);
  end_t=clock();
  cpu_time_used=((double)(end_t-start_t))/CLOCKS_PER_SEC;
  printf("\nfactorial of %d is %d using iteration & it take %f seconds..",n,ite_result,cpu_time_used);
  return 0;
}
```

```python
%%writefile copilot.c
#include<stdio.h>
int fact(int current,int n,int result){
  if(current>n){
    return result;
    }
  return fact(current+1,n,result*current);
}
int main(){
  int n=5;
  printf("factorial of %d is %d",n,fact(1,n,1));
  return 0;
}
```

```python
%%writefile prof.c
#include<stdio.h>
#include<time.h>

//Top-down recursive approach
int td_fact(int n){
  if(n==0   || n==1){
    return 1;
  }
  else{
    return n*td_fact(n-1);
  }
}

//Bottom-up recursive approach
int bu_fact(int current,int n,int result){
  if(current>n){
    return result;
  }
  return bu_fact(current+1,n,result*current);
}

int main()
{
  int n=5;
  clock_t start, end;
  double cpu_time_used;

  //profiling of top-down recusive approach..
  start=clock();
  int td_result=td_fact(n);
  end=clock();
  cpu_time_used=((double)(end-start))/CLOCKS_PER_SEC;
  printf("factorial of %d is %d",n,td_result);
  printf("\n Time consumed : %f",cpu_time_used);

  //profiling of bottom-up recusive approach..
  start=clock();
  int bu_result=bu_fact(1,n,1);
  end=clock();
  cpu_time_used=((double)(end-start))/CLOCKS_PER_SEC;
  printf("\nfactorial of %d is %d",n,bu_result);
  printf("\n Time consumed : %f",cpu_time_used);

  return 0;
}
```
